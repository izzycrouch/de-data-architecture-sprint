Ingest the Data

```python
import pandas as pd
import csv

# Get customer raw data from landing zone
with open('./landing_zone/customer_data.csv', encoding='utf-8') as c:
    customer_data = csv.DictReader(c)

customer_df = pd.DataFrame.from_dict(customer_data)
customer_df
```

```python
# Get order raw data from landing zone
with open('./landing_zone/order_data.csv', encoding='utf-8') as o:
    order_data = csv.DictReader(o)

order_df = pd.DataFrame.from_dict(order_data)
order_df
```

```python
# Get restaurants raw data from landing zone
with open('./landing_zone/restaurants_data.csv', encoding='utf-8') as r:
    restaurant_data = csv.DictReader(r)

restaurant_df = pd.DataFrame.from_dict(restaurant_data)
restaurant_df
```
