```python
import csv
import random
from datetime import datetime, timedelta
import pandas as pd

def generate_raw_data(num_records=10):
    restaurants = ["Pizza  Time", "Sushi  Go", "Burger Hub", "Taco Town"]
    customers = ["Alice B", "Bob c", "Charlie An", "Diana U", "Ethan  Y",
                "Fiona H", "George k", "Hannah  Bn", "Iris TY", "Jack GH"]
    area = ['Manchester', 'Leeds', 'London', 'Hull']
    address = ['AA-AA', 'CC cc', 'ff-ff']

    rows = []
    for i in range(1, num_records +1):
    
        data = {
            "order_num": (i + 2)*2,
            "restaurant": random.choice(restaurants),
            "area": random.choice(area),
            "customer_id":i + 1,
            "customer_name": random.choice(customers),
            "customer_address": random.choice(address),
            "data": (datetime.now() - timedelta(days=random.randint(0, 30))).strftime("%Y-%m-%d %H:%M:%S"),         
            "quanitty_of_items": random.randint(2,7),
            "cost": round(random.uniform(10, 50), 2)
    }   
    
        rows.append(data)


    fieldnames = ["order_num", "restaurant", "area", "customer_id", "customer_name", 
              "customer_address", "data", "quanitty_of_items", "cost"]

    with open('/Users/yana/northcoders/Week8/de-data-architecture-sprint/landing_zone/customer_data.csv', 'w', newline="", encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()    
        writer.writerows(rows)

customer_data = generate_raw_data(15)
    
```
